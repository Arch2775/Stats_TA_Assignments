# **Mean, Variance, and Standard Deviation**

Mean: The mean (average) is the sum of all values in a dataset divided by the number of values. It represents the central point of the data.
image.png

Variance: Variance measures the average degree to which each point differs from the mean. It shows how spread out the data points are.

image.png

Standard Deviation: The standard deviation is the square root of the variance. It gives a measure of the spread of the data points around the mean in the same units as the data.

image.png

# **5-Number Summary (Boxplots)**

The 5-number summary provides a quick overview of the distribution of a dataset and is often visualized using a boxplot. It consists of the following:


1.**Minimum**: The smallest data point in the dataset.

2.**First Quartile (Q1)**: The 25th percentile, where 25% of the data points lie below this value.

3.**Median (Q2)**: The 50th percentile, representing the middle value of the dataset.

4.**Third Quartile (Q3)**: The 75th percentile, where 75% of the data points lie below this value.

5.**Maximum**: The largest data point in the dataset.

**Boxplot**: A boxplot visually represents the 5-number summary, showing the median, quartiles, and potential outliers. The box indicates the interquartile range (IQR), which is the spread between Q1 and Q3, while the whiskers extend to the minimum and maximum values within 1.5 times the IQR.



These statistical measures help summarize the data's central tendency, spread, and overall distribution, providing insights into patterns and outliers in the dataset.
